## Record function

```python
from IPython.display import Javascript
from google.colab import output
from base64 import b64decode
from io import BytesIO
!pip -q install pydub
from pydub import AudioSegment
```

```python
RECORD = """
const sleep  = time => new Promise(resolve => setTimeout(resolve, time))
const b2text = blob => new Promise(resolve => {
  const reader = new FileReader()
  reader.onloadend = e => resolve(e.srcElement.result)
  reader.readAsDataURL(blob)
})
var record = time => new Promise(async resolve => {
  stream = await navigator.mediaDevices.getUserMedia({ audio: true })
  recorder = new MediaRecorder(stream)
  chunks = []
  recorder.ondataavailable = e => chunks.push(e.data)
  recorder.start()
  await sleep(time)
  recorder.onstop = async ()=>{
    blob = new Blob(chunks)
    text = await b2text(blob)
    resolve(text)
  }
  recorder.stop()
})
"""

def record(sec=3):
  display(Javascript(RECORD))
  s = output.eval_js('record(%d)' % (sec*1000))
  b = b64decode(s.split(',')[1])
  with open('audio.wav','wb') as f:
    f.write(b)
  return 'audio.wav'
```

## Webpage opener

```python
def open_web(url):
    display(Javascript('window.open("{url}");'.format(url=url)))
```

```python
keys = ['score', 'schedule', 'maya', 'course', 'homework', 'notification', 'finance', 'forum', 'lingua', 'inbox']
values = ['https://binusmaya.binus.ac.id/newStudent/#/score/viewscore',
          'https://newbinusmaya.binus.ac.id/lms/schedule',
          'https://newbinusmaya.binus.ac.id/lms/dashboard',
          'https://newbinusmaya.binus.ac.id/lms/course',
          'https://newbinusmaya.binus.ac.id/lms/assessment',
          'https://binusmaya.binus.ac.id/newGeneral/#/notification',
          'https://binusmaya.binus.ac.id/newStudent/#/financial/financialSummary',
          'https://newbinusmaya.binus.ac.id/lms/forum',
          'https://newbinusmaya.binus.ac.id/beelingua/',
          'https://binusmaya.binus.ac.id/newGeneral/#/message/inbox']

fitur = {keys[i]: values[i] for i in range(len(keys))}
```

## Activation function

```python
def activate():

  print('Recording...')

  audio = record(5)

  print('Transcribing recorded speech:')

  transcript = model.transcribe(['/content/audio.wav'])

  print(f"\n\nTranscribed recorded speech: {transcript[0]['transcription']}")

  for key in fitur:

    if str(key) in transcript[0]['transcription']:

      print(f"\nOpening webpage: {str(fitur[key])}")

      return open_web(str(fitur[key]))

  return print("\nError: Command not recognized, please try again.")
```
